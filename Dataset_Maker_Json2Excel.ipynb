```python
def LoadDataToExelSheets(_jsonFormatStringDataFileName):
  # Row Count, Data Start Index = 2 
  exel_row_count = 2

  # Create Workbook
  wb = Workbook()
  ws = wb.active
  ws.title = 'ETC_DataSet'

  
  bold_font = Font(bold= True)
  
  # Resize column width for comfort view
  ws.column_dimensions['A'].width = 5
  
  for i in ['B','C','D','E']:
    ws.column_dimensions[i].width = 25
    ws[i+str(1)].alignment = Alignment(horizontal='center')
    ws[i+str(1)].font = bold_font

  for i in ['F','G','H','I','J','K','L']:
    ws.column_dimensions[i].width = 10
    ws[i+str(1)].alignment = Alignment(horizontal='center')
    ws[i+str(1)].font = bold_font

  # Features
  ws['B1'] = "candle_acc_trade_price"
  ws['C1'] = "candle_acc_trade_volume"
  ws['D1'] = "candle_date_time_kst"
  ws['E1'] = "candle_date_time_utc"
  ws['F1'] = 'high_price'
  ws['G1'] = 'low_price'
  ws['H1'] = 'market'
  ws['I1'] = 'opening_price'
  ws['J1'] = 'timestamp'
  ws['K1'] = 'trade_price'
  ws['L1'] = 'unit'

  # Read json format string file 
  f = open(_jsonFormatStringDataFileName, "r")
  
  while(True):
    # Read line by line
    line = f.readline()

    # exit loop when EOF
    if not line: break
    data = line.rstrip().split(" ")

    index = data[0]
    json_data = json.loads(data[1])
    
    # Alignment setting
    for i in ['B','C','D','E','F','G','H','I','J','K','L']:    
      ws[i+str(exel_row_count)].alignment = Alignment(horizontal='center')

    ws['A'+str(exel_row_count)] = index
    ws['B'+str(exel_row_count)] = json_data[0]['candle_acc_trade_price']
    ws['C'+str(exel_row_count)] = json_data[0]['candle_acc_trade_volume']
    ws['D'+str(exel_row_count)] = json_data[0]['candle_date_time_kst']
    ws['E'+str(exel_row_count)] = json_data[0]['candle_date_time_utc']
    ws['F'+str(exel_row_count)] = json_data[0]['high_price']
    ws['G'+str(exel_row_count)] = json_data[0]['low_price']
    ws['H'+str(exel_row_count)] = json_data[0]['market']
    ws['I'+str(exel_row_count)] = json_data[0]['opening_price']
    ws['J'+str(exel_row_count)] = json_data[0]['timestamp']
    ws['K'+str(exel_row_count)] = json_data[0]['trade_price']
    ws['L'+str(exel_row_count)] = json_data[0]['unit']

    # Increment row count for next Data
    exel_row_count = exel_row_count + 1

  # Save Workbook file and close
  wb.save('ETC_DataSet.xlsx')
  wb.close()
```
